```python
merged_data['Na'] = merged_data.groupby(['病歷號', '收件日期'])['Na'].transform(lambda x: x.fillna(method='bfill'))
merged_data['K'] = merged_data.groupby(['病歷號', '收件日期'])['K'].transform(lambda x: x.fillna(method='bfill'))
merged_data['Mg'] = merged_data.groupby(['病歷號', '收件日期'])['Mg'].transform(lambda x: x.fillna(method='bfill'))
merged_data['Cl'] = merged_data.groupby(['病歷號', '收件日期'])['Cl'].transform(lambda x: x.fillna(method='bfill'))
merged_data['BUN (B)'] = merged_data.groupby(['病歷號', '收件日期'])['BUN (B)'].transform(lambda x: x.fillna(method='bfill'))
merged_data['Creatinine(B)'] = merged_data.groupby(['病歷號', '收件日期'])['Creatinine(B)'].transform(lambda x: x.fillna(method='bfill'))
merged_data.head()
```

```python
merged_data_dropna = merged_data.dropna(subset=['Temperature', 'pH', 'pCO2', 'pO2', 'HCO3', 'TCO2', 'ABE', 'SBE', 'SBC', 'SAT'], how='all')
merged_data_dropna.shape
```

```python
merged_data_tco2 = merged_data_dropna['TCO2'].value_counts().sort_index().to_frame()
merged_data_tco2
```

```python
merged_data_tco2.index[0]
```

```python
merged_data_dropna['TCO2'].replace({'--------': None}, inplace=True)
merged_data_dropna['TCO2'] = pd.to_numeric(merged_data_dropna['TCO2'])
merged_data_dropna.info()
```

```python
merged_data_bun_b = merged_data_dropna['BUN (B)'].value_counts().sort_index().to_frame()
merged_data_bun_b
```

```python
merged_data_bun_b.index[0]
```

```python
merged_data_dropna['BUN (B)'].replace({'     < 1': 1}, inplace=True)
merged_data_dropna['BUN (B)'] = pd.to_numeric(merged_data_dropna['BUN (B)'])
merged_data_dropna.info()
```

```python
merged_data_creatinine_b = merged_data_dropna['Creatinine(B)'].value_counts().sort_index().to_frame()
merged_data_creatinine_b
```

```python
merged_data_creatinine_b.index[0]
```

```python
merged_data_dropna['Creatinine(B)'].replace({'   < 0.1': 1}, inplace=True)
merged_data_dropna['Creatinine(B)'] = pd.to_numeric(merged_data_dropna['Creatinine(B)'])
merged_data_dropna.info()
```

```python
merged_data_fio2 = merged_data_dropna['fio2'].value_counts().sort_index().to_frame()
merged_data_fio2
```

```python
merged_data_dropna['fio2'].replace({'35%': 35}, inplace=True)
merged_data_dropna['fio2'].replace({'50%': 50}, inplace=True)
merged_data_dropna['fio2'] = pd.to_numeric(merged_data_dropna['fio2'])
merged_data_dropna.info()
```

```python
merged_data_tidal_volume = merged_data_dropna['tidal volume'].value_counts().sort_index().to_frame()
merged_data_tidal_volume
```

```python
merged_data_dropna['tidal volume'].replace({'-': None}, inplace=True)
merged_data_dropna['tidal volume'] = pd.to_numeric(merged_data_dropna['tidal volume'])
merged_data_dropna.info()
```

```python
merged_data_respiratory_rate = merged_data_dropna['respiratory rate'].value_counts().sort_index().to_frame()
merged_data_respiratory_rate
```

```python
merged_data_dropna['respiratory rate'] = pd.to_numeric(merged_data_dropna['respiratory rate'])
merged_data_dropna.info()
```

```python
merged_data_minute_ventilation = merged_data_dropna['minute ventilation'].value_counts().sort_index().to_frame()
merged_data_minute_ventilation
```

```python
merged_data_dropna['minute ventilation'].replace({'443/': 4.43}, inplace=True)
merged_data_dropna['minute ventilation'] = pd.to_numeric(merged_data_dropna['minute ventilation'])
merged_data_dropna.info()
```

```python
merged_data_spo2 = merged_data_dropna['spo2'].value_counts().sort_index().to_frame()
merged_data_spo2
```

```python
merged_data_dropna['spo2'].replace({'88-93': 90.5}, inplace=True)
merged_data_dropna['spo2'].replace({'91-96': 93.5}, inplace=True)
merged_data_dropna['spo2'].replace({'92-94': 93}, inplace=True)
merged_data_dropna['spo2'].replace({'POOR': 50}, inplace=True)
merged_data_dropna['spo2'].replace({'poor': 50}, inplace=True)
merged_data_dropna['spo2'] = pd.to_numeric(merged_data_dropna['spo2'])
merged_data_dropna.info()
```
